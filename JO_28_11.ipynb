```python
import pandas as pd

df = pd.read_csv('titanic.csv')
print(df.head())
```

```python
df2 = df[['Name', 'Age', 'Sex']]
print(df2.head())
```

```python
median_age = df2['Age'].median()
print(f"The median age of the passengers is: {median_age}")
```

```python
df2['Sex'] = df2['Sex'].replace({'male': 0, 'female': 1})
print(df2.head())

women_count = df2[df2['Sex'] == 1].shape[0]
men_count = df2[df2['Sex'] == 0].shape[0]

print(f"Number of women on board: {women_count}")
print(f"Number of men on board: {men_count}")
```
